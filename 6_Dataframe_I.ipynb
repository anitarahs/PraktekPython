## Creating New Variable/Kolom
1. Agregasi (tambahkan price dan freight_value)
2. Agregasi (lakukan pangkat 2 pada freight_value)
3. Concantenate (gabungin product_id dan seller_id) 

```python
# intermezo
conn = psycopg2.connect(
          host = "digitalskoladb.c04me33o8tni.ap-southeast-1.rds.amazonaws.com",
          database = 'sandbox',
          user = 'ds20_all',
          password = 'dsBatch20')
conn.set_client_encoding('UTF8')


# reading data from sql
sql_order_itm = """select *
  from batch_16.order_items_dataset;"""
order_item_data = pd.read_sql_query(sql_order_itm, conn)
order_item_data = order_item_data.reset_index(drop=True)
order_item_data.head()
```

```python
# 1. Agregasi (tambahkan price dan freight_value)
order_item_data['price_&_freigt_value'] = order_item_data['price']+order_item_data['freight_value']
order_item_data.head()
```

```python
# 2. Agregasi (lakukan pangkat 2 pada freight_value)
order_item_data['freigt_value_pangkat'] = order_item_data['freight_value']**2
order_item_data.head()
```

```python
# 3. Concantenate (gabungin product_id dan seller_id)
order_item_data['product_seller_id'] = order_item_data['product_id']+str("-")+order_item_data['seller_id']
order_item_data.head()
```

```python
np.dtype(order_item_data['shipping_limit_date'])
```

## Grouping Data

1. Ambil rata-rata dan median dari price per product_id
2. Ambil nilai maximum dan minimum dari price per product_id (min and max)
3. Ambil jumlah product yang dijumlah di tiap seller_id (count)
4. Berapa banyak jumlah customer pada customer_city dan customer_state (count)

```python
# 1. Ambil rata-rata dan median dari price
agregasi_price = order_item_data.groupby('product_id').agg(average_price = ('price','mean'), median_price = ('price','median')).reset_index() #reset_indeks untuk membuat data menjadi dalam bentuk indeks dataframe
agregasi_price
```

```python
#jumlah semua berdasarkan total per product_id
order_item_data.groupby('product_id').agg('count')
```

```python
order_item_data[order_item_data['product_id']=='fff6177642830a9a94a0f2cba5e476d1']
```

```python
# 1. Ambil sample 20 data nilai rata-rata dan median dari price
agregasi_price = order_item_data.groupby('product_id').agg(average_price = ('price','mean'), median_price = ('price','median')).reset_index()
agregasi_price.sample(20)
```

```python
# 2. Ambil nilai maximum dan minimum dari price per product_id (min and max)
agregasi_price1 = order_item_data.groupby('product_id').agg(min_price = ('price','min'), max_price = ('price','max')).reset_index()
agregasi_price1.head()
```

```python
order_item_data.groupby('product_id').agg(max_price = ('price','max'), min_price = ('price','min')).reset_index()
```

```python
#tanpa nama membuat nama kolom baru
agregasi_price = order_item_data.groupby('product_id').agg({'order_item_id':'count','price':'mean'}).reset_index()
agregasi_price
```

```python
# 3. amount of product per seller_id (jumlah produk per seller_id)
aggregasi_ = order_item_data.groupby('seller_id').agg(amount_of_product = ('product_id', 'count')).reset_index()
aggregasi_
```
